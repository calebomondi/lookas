## OPENCODE

```python
models = [
    "qwen3.7-plus",
    "deepseek-v4-pro"
]

try:
    for model in models:
        result = _llm("opencode", model).invoke("What is the capital of France?")
        print(f"OpenCode/{model}: {result}")
except Exception as e:
    print(f"OpenCode Error: {e}")
```
